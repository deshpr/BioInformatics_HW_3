## 2. Linear Regression.

```r
lmFit1 <- lm (my.data$CDK1 ~ my.data$UBE2C)
```

```r
summary (lmFit1)
```

```r
# maximum correlation computed last Wed was 0.5227213
sqrt (0.2732)
```

```r
names (lmFit1)
```

```r
lmFit1$coefficients
```

```r
coef (lmFit1)
```

```r
# compute predicted values given these estimated regression coefficients
predict.lmFit1 <- predict (lmFit1)
```

```r
predict.lmFit1
```

```r
curr.RSS <- sum ((my.data$CDK1 - predict.lmFit1)^2)
```

```r
curr.RSS
```

```r
curr.RSE <- sqrt (curr.RSS/(438-2))
curr.RSE
```

```r
# compute R squared ourselves
curr.TSS <- sum ((my.data$CDK1 - mean (my.data$CDK1))^2)
curr.Rsq <- (curr.TSS - curr.RSS)/curr.TSS
curr.Rsq
```

```r
plot (my.data$UBE2C, my.data$CDK1)
abline (lmFit1)
```
